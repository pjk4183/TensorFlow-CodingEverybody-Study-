```python
import pandas as pd

# 파일 읽어오기: pd.read_csv('/경로/파일명.csv')
# 모양 확인하기: print(데이터.shape)
# 칼럼 선택하기: 데이터[['칼럼명1', '칼럼명2', '칼럼명3']]
# 칼럼 이름 출력하기: print(데이터.columns)
# 맨 위 5개 관측치 출력하기: 데이터.head()
```

```python
#reading files
boston = pd.read_csv('boston.csv')
lemonade = pd.read_csv('lemonade.csv')
iris = pd.read_csv('iris.csv')

#check data shapes (row, column)
print(boston.shape)
print(lemonade.shape)
print(iris.shape)

#columns name
print(boston.columns)
print(lemonade.columns)
print(iris.columns)
```

```python
독립 = lemonade[['온도']]
종속 = lemonade[['판매량']]
print(독립.shape, 종속.shape)

독립 = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
종속 = boston[['medv']]
print(독립.shape, 종속.shape)

독립 = iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = iris[['품종']]
print(독립.shape, 종속.shape)
```
